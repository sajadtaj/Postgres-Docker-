## Call packages

```python
import psycopg2
import pandas as pd
```

## Establishing the connection

```python
#Establishing the connection
conn = psycopg2.connect(
    database="dvdrental", user='admin', password='root', host='127.0.0.1', port= '5432'
)
#Setting auto commit false
conn.autocommit = True


#Creating a cursor object using the cursor() method
cursor = conn.cursor()
```

## Select

### *

```python
# Preparing SQL queries to INSERT a record into the database.
cursor.execute('''
SELECT * FROM customer;
      ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names)

# Displaying the results
for row in results:
      print(row)  # Print the first_name from each row
```

### Aliases & ||

```python
# Preparing SQL queries to INSERT a record into the database.
cursor.execute('''
SELECT first_name || ' ' || last_name as fullname
FROM customer;
      ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names[0])
print("---"*5)

# Displaying the results
for row in results:
      print(row[0])  # Print the first_name from each row
```

### ORDER BY 

```python
# Preparing SQL queries to INSERT a record into the database.
cursor.execute('''
    SELECT 
        first_name, 
        last_name 
    FROM 
        customer 
    ORDER BY 
        first_name ASC;
        ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names)
print("---"*5)

# Displaying the results
for row in results:
    print(row)  # Print the first_name from each row
```

```python
cursor.execute('''
    SELECT 
        first_name, 
        last_name 
    FROM 
        customer 
    ORDER BY 
        first_name ASC,
        last_name DESC;
        ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names)
print("---"*5)

# Displaying the results
for row in results:
    print(row)  # Print the first_name from each row
```

### LENGTH()

```python
cursor.execute('''
    SELECT 
        first_name, 
        LENGTH(last_name) as length 
    FROM 
        customer 
    ORDER BY 
        length DESC;
        ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names)
print("---"*5)

# Displaying the results
for row in results:
    print(row)  # Print the first_name from each row
```

###   NULL
+   ORDER BY 
+   sort_expresssion [ASC | DESC] [NULLS FIRST | NULLS LAST]

The NULLS FIRST option places NULL before other non-null values and the NULL LAST option places NULL after other non-null values.


```python
cursor.execute('''
    SELECT 
        first_name
    FROM 
        customer 
    ORDER BY 
        first_name DESC NULLS FIRST;
        ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names)
print("---"*5)

# Displaying the results
for row in results:
    print(row)  # Print the first_name from each row
```

## DISTINCT
+   SELECT 
+     DISTINCT column1 
+   FROM 
+     table_name;

The SELECT DISTINCT removes duplicate rows from a result set. The SELECT DISTINCT clause retains one row for each group of duplicates.

```python
cursor.execute('''
    SELECT 
        distinct first_name 
    FROM 
        customer 
        ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names)
print("---"*5)

# Displaying the results
for row in results:
    print(row)  # Print the first_name from each row
```

```python
cursor.execute('''
    SELECT 
        distinct * 
    FROM 
        customer 
        ''')


# Fetching the results
results = cursor.fetchall()

# Column Name
column_names = [desc[0] for desc in cursor.description]

# Displaying the column names
print(column_names)
print("---"*5)

# Displaying the results
for row in results:
    print(row)  # Print the first_name from each row
```
